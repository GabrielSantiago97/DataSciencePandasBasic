##IMPORTAR DADOS



```python
from google.colab import files
uploaded = files.upload()
```

```python
import io
import pandas as pd
import matplotlib.pyplot as plt
base = pd.read_csv(io.BytesIO(uploaded['best-selling-nintendo64.csv']))
```

##ANALISAR DADOS

```python
date = pd.DatetimeIndex(base['Release date'])
base['Ano'] = date.year
base['Mes'] = date.month
base['Dia'] = date.day
```

```python
base = base.dropna()
```

```python
base.head()
```

```python
base.info()
```

```python
#TOP 10 JOGOS MAIS VENDIDOS
jogo = base[["Game", "Sales"]].groupby("Game").sum()
jogo.sort_values(by="Sales", ascending=False).head(10)
```

```python
#GRÁFICO TOP 5 JOGOS MAIS VENDIDOS
jogo = base[["Game", "Sales"]].groupby("Game").sum()
jogo.sort_values(by="Sales", ascending=False).head().plot.bar()
```

```python
#GRÁFICO VENDAS POR MESES
jogo = base[["Mes", "Sales"]].groupby("Mes").sum()
jogo.sort_values(by="Sales", ascending=False).plot.bar()
```

```python
#GRÁFICO VENDAS POR ANO
jogo = base[["Ano", "Sales"]].groupby("Ano").sum()
jogo.sort_values(by="Sales", ascending=False).plot.bar()
```

```python
#RANK VENDAS POR DEVELOPERS 
jogo = base[["Developer(s)", "Sales"]].groupby("Developer(s)").sum()
jogo.sort_values(by="Sales", ascending=False)
```
